BMI Calculator

```python
# Import necessary libraries
import math

# Function to convert height units
def height_converter():
    print("\nHeight Conversion Options:")
    print("1. Centimeters to Meters")
    print("2. Feet and Inches to Meters")
    choice = int(input("Select an option (1 or 2): "))

    if choice == 1:
        cm = float(input("Enter height in centimeters: "))
        return cm / 100  # Convert cm to meters

    elif choice == 2:
        feet = float(input("Enter feet: "))
        inches = float(input("Enter inches: "))
        return (feet * 0.3048) + (inches * 0.0254)  # Convert to meters

    else:
        print("Invalid choice! Defaulting to meters.")
        return float(input("Enter height in meters: "))

# Function to convert weight units
def weight_converter():
    print("\nWeight Conversion Options:")
    print("1. Kilograms")
    print("2. Pounds to Kilograms")
    choice = int(input("Select an option (1 or 2): "))

    if choice == 1:
        return float(input("Enter weight in kilograms: "))

    elif choice == 2:
        pounds = float(input("Enter weight in pounds: "))
        return pounds * 0.453592  # Convert pounds to kilograms

    else:
        print("Invalid choice! Defaulting to kilograms.")
        return float(input("Enter weight in kilograms: "))

# Function to calculate BMI
def calculate_bmi(weight, height):
    return weight / (height ** 2)

# Function to categorize BMI with obesity levels
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    elif 30 <= bmi < 34.9:
        return "Obesity Class 1"
    elif 35 <= bmi < 39.9:
        return "Obesity Class 2"
    else:
        return "Obesity Class 3 (Severe)"

# Function to provide a detailed diet plan
def suggest_diet_plan(category, conditions):
    print("\nDiet Recommendations:")

    if category == "Underweight":
        print("- **Focus on calorie-dense foods**: Nuts, seeds, avocados, and whole grains.")
        print("- **Increase protein intake**: Eggs, lean meats, dairy, tofu, and legumes.")
        print("- **Frequent meals**: Eat every 3-4 hours with snacks in between.")
        print("- **Include healthy fats**: Olive oil, peanut butter, and fatty fish.")

    elif category == "Normal weight":
        print("- **Balanced meals**: Include vegetables, lean protein, and complex carbs.")
        print("- **Portion control**: Avoid overeating and use smaller plates.")
        print("- **Hydration**: Drink at least 2-3 liters of water daily.")
        print("- **Moderate treats**: Enjoy sweets occasionally but in moderation.")

    elif category == "Overweight":
        print("- **Calorie-deficit diet**: Aim to consume 500-700 calories less than maintenance.")
        print("- **Avoid processed carbs**: Replace white bread and pasta with whole grains.")
        print("- **Increase fiber intake**: Vegetables, fruits, oats, and chia seeds.")
        print("- **Meal prep**: Plan meals in advance to avoid impulsive eating.")

    elif category == "Obesity Class 1":
        print("- **Low-GI foods**: Lentils, quinoa, and whole wheat to regulate blood sugar.")
        print("- **Smaller meals**: Eat 5-6 small meals a day to prevent hunger spikes.")
        print("- **Limit sugary beverages**: Replace with green tea or infused water.")
        print("- **Protein-rich breakfast**: Helps control cravings throughout the day.")

    elif category == "Obesity Class 2":
        print("- **Follow a low-carb diet**: Avoid starchy foods like potatoes and rice.")
        print("- **Increase plant-based foods**: Salads, smoothies, and stir-fried vegetables.")
        print("- **Practice mindful eating**: Chew slowly and avoid distractions during meals.")
        print("- **Hydration is key**: Drink a glass of water 30 minutes before meals.")

    elif category == "Obesity Class 3 (Severe)":
        print("- **Medical supervision recommended**: Consult a dietitian for a personalized plan.")
        print("- **Meal replacement shakes**: Use as part of a monitored weight-loss program.")
        print("- **Low-impact foods**: Soups, lean protein, and steamed vegetables.")
        print("- **Monitor salt intake**: Avoid high-sodium foods to prevent water retention.")

    if "PCOS" in conditions:
        print("\n**PCOS-specific advice:**")
        print("- Avoid high-carb foods and sugars.")
        print("- Focus on anti-inflammatory foods: Turmeric, berries, and leafy greens.")
        print("- Incorporate strength training to improve insulin sensitivity.")

    if "Diabetes" in conditions:
        print("\n**Diabetes-specific advice:**")
        print("- Monitor carbohydrate intake and stick to low-GI foods.")
        print("- Avoid sugary snacks and opt for whole fruits.")
        print("- Include lean protein to slow digestion and prevent sugar spikes.")

    if "Hypertension" in conditions:
        print("\n**Hypertension-specific advice:**")
        print("- Follow a DASH diet: Rich in fruits, vegetables, and low-fat dairy.")
        print("- Avoid processed foods and reduce salt intake.")
        print("- Increase potassium intake: Bananas, spinach, and sweet potatoes.")

# Function to provide a personalized exercise plan
def suggest_exercise_plan(category):
    print("\nExercise Recommendations:")

    if category in ["Underweight", "Normal weight"]:
        print("- **Moderate exercise**: 30 minutes of brisk walking or cycling daily.")
        print("- **Strength training**: 2-3 times a week to build or maintain muscle.")

    elif category == "Overweight":
        print("- **Cardio workouts**: Swimming, running, or aerobics 3-5 times a week.")
        print("- **Strength training**: To improve metabolism and burn fat.")

    elif category == "Obesity Class 1":
        print("- **Low-impact activities**: Yoga, walking, or water aerobics.")
        print("- **Gradually increase intensity** as fitness improves.")

    elif category in ["Obesity Class 2", "Obesity Class 3 (Severe)"]:
        print("- **Supervised physical therapy** or trainer-led sessions.")
        print("- **Focus on flexibility and balance** to avoid injury.")
        print("- **Walking or water therapy** to reduce joint stress.")

# Main function
def main():
    print("Welcome to the Enhanced BMI Calculator!")
    print("Created by Ruba Fareed")

    # Collecting user information
    age = int(input("Enter your age: "))
    gender = input("Enter your gender (Man/Woman/Child): ").capitalize()

    if gender == "Woman":
        pregnant = input("Are you pregnant? (yes/no): ").strip().lower()
        if pregnant == 'yes':
            print("Pregnancy detected. Please consult your doctor for specific advice.")

    # Collect health conditions
    conditions = []
    print("\nDo you have any of the following conditions? (yes/no)")
    if input("Diabetes: ").strip().lower() == "yes":
        conditions.append("Diabetes")
    if input("Hypertension: ").strip().lower() == "yes":
        conditions.append("Hypertension")
    if input("Hypotension: ").strip().lower() == "yes":
        conditions.append("Hypotension")
    if input("PCOS (Polycystic Ovary Syndrome): ").strip().lower() == "yes":
        conditions.append("PCOS")

    # Get height and weight through converter functions
    height = height_converter()
    weight = weight_converter()

    # Calculate and categorize BMI
    bmi = calculate_bmi(weight, height)
    category = categorize_bmi(bmi)

    # Provide diet and exercise plan
    suggest_diet_plan(category, conditions)
    suggest_exercise_plan(category)

# Run the program
if __name__ == "__main__":
    main()
```
